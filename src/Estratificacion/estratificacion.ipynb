```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold

# Ordenar primero por precio
data_min.sort_values(by='min_price', inplace=True)

# Agrupar por artista
grouped = data_min.groupby('nameArtist')

# Inicializar el objeto KFold
n_splits = 4
kf = KFold(n_splits=n_splits)

# Inicializar una lista para almacenar las filas asignadas a cada fold
filas_por_fold = [[] for _ in range(n_splits)]

# Iterar sobre los grupos de artista y distribuirlos equitativamente en los pliegues
for _, group in grouped:
    # Calcular el tamaño del grupo y el número de conciertos por pliegue
    group_size = len(group)
    fold_size = group_size // n_splits
    remainder = group_size % n_splits

    # Repartir los conciertos del grupo en los pliegues
    start = 0
    for fold, fold_count in enumerate(np.full(n_splits, fold_size)):
        if remainder > 0:
            fold_count += 1
            remainder -= 1

        end = start + fold_count
        filas_por_fold[fold].extend(group.iloc[start:end].index)
        start = end

# Imprimir el número de conciertos asignados a cada fold
for fold, filas in enumerate(filas_por_fold):
    print(f"Número de conciertos en el fold {fold + 1}:", len(filas))

# Ahora puedes acceder a las filas asignadas a cada fold utilizando la lista filas_por_fold
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold

# Agrupamos primero por nombre y después ordenamos por precio
data_min = data_min.sort_values(by=['name', 'min_price'])

# Reindexar para evitar problemas con índices después de la ordenación
data_min = data_min.reset_index(drop=True)

# Calcula el número total de conciertos
total_conciertos = len(data_min)

# Define el número de folds
n_folds = 4

# Calcula el tamaño de fold aproximado
tamaño_fold_aproximado = total_conciertos // n_folds

# Calcula el tamaño real de cada fold (ajustado para que sea lo más equitativo posible)
tamaño_fold_ajustado = [tamaño_fold_aproximado] * n_folds
resto = total_conciertos % n_folds
for i in range(resto):
    tamaño_fold_ajustado[i] += 1

# Inicializa el objeto KFold
kf = KFold(n_splits=n_folds)

# Distribuye los conciertos en los folds
fold_assignments = np.zeros(total_conciertos, dtype=int)
start_index = 0
for fold, tamaño in enumerate(tamaño_fold_ajustado):
    end_index = start_index + tamaño
    fold_assignments[start_index:end_index] = fold
    start_index = end_index

# Asignar los folds al DataFrame
data_min['fold'] = fold_assignments

# Calcular el número de conciertos asignados a cada fold
conciertos_por_fold = [np.sum(fold_assignments == fold) for fold in range(n_folds)]

# Imprimir el número de conciertos asignados a cada fold
for fold, conciertos in enumerate(conciertos_por_fold, start=1):
    print(f"Número de conciertos en el fold {fold}: {conciertos}")

```
